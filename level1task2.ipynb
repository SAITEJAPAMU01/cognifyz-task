```python
import pandas as pd

# Load the Excel file
df = pd.read_excel(r'C:\Users\saite\OneDrive\Desktop\INTERN.xlsx')


numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns


for column in numerical_columns:
    print(f"Statistics for {column}:")
    print(f"Mean: {df[column].mean()}")
    print(f"Median: {df[column].median()}")
    print(f"Mode: {df[column].mode()[0]}")
    print(f"Standard Deviation: {df[column].std()}")
    print(f"Variance: {df[column].var()}")
    print(f"Minimum: {df[column].min()}")
    print(f"Maximum: {df[column].max()}")
    print(f"Range: {df[column].max() - df[column].min()}")
    print(f"Quartiles:\n{df[column].quantile([0.25, 0.5, 0.75])}")
    print(f"Interquartile Range (IQR): {df[column].quantile(0.75) - df[column].quantile(0.25)}")
    print("\n")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file
df = pd.read_excel(r'C:\Users\saite\OneDrive\Desktop\INTERN.xlsx')

# "Country Code"
country_counts = df['Country Code'].value_counts()
print('Distribution of Country Code:')
print(country_counts)

# distribution of "Country Code"
plt.figure(figsize=(10, 6))
sns.barplot(x=country_counts.index, y=country_counts.values)
plt.title('Distribution of Country Code')
plt.xlabel('Country Code')
plt.ylabel('Count')
plt.show()

#  the distribution of "City"
city_counts = df['City'].value_counts()
print('Distribution of City:')
print(city_counts)

# Plot the distribution of "City"
plt.figure(figsize=(12, 8))
sns.barplot(x=city_counts.index, y=city_counts.values)
plt.title('Distribution of City')
plt.xlabel('City')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()

# Explore the distribution of "Cuisines"
df['Cuisines'] = df['Cuisines'].str.split(', ')
cuisine_counts = df['Cuisines'].explode().value_counts()
print('Distribution of Cuisines:')
print(cuisine_counts)

# Plot the distribution of "Cuisines"
plt.figure(figsize=(12, 8))
sns.barplot(x=cuisine_counts.index, y=cuisine_counts.values)
plt.title('Distribution of Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()

# Identify the top cuisines and cities with the highest number of restaurants
top_cuisines = cuisine_counts.head(10)
top_cities = city_counts.head(10)

print('Top 10 Cuisines:')
print(top_cuisines)

print('Top 10 Cities:')
print(top_cities)

#  top 10 cuisines and cities
plt.figure(figsize=(12, 6))
sns.barplot(x=top_cuisines.index, y=top_cuisines.values)
plt.title('Top 10 Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x=top_cities.index, y=top_cities.values)
plt.title('Top 10 Cities with Highest Number of Restaurants')
plt.xlabel('City')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```
